#### Задание 2. Есть ли категориальные признаки? Что с ними делать?

Категориальные признаки кодируем посредством get_dummies

#### Задание 3. Фунция предобработки

Напишите функцию, которая бы

* Удаляло идентификатор `AGREEMENT_RK`
* Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
* Что-то делала с пропусками
* Кодировала категориальные признаки

В результате, ваш датафрейм должен содержать только числа и не содержать пропусков!

```python
def preproc_data(df_input):
    df_output = df_input.copy()
    
    df_output = df_output.drop('AGREEMENT_RK', axis=1)
    for column in 'PERSONAL_INCOME,CREDIT,FST_PAYMENT,LOAN_AVG_DLQ_AMT,LOAN_MAX_DLQ_AMT'.split(','):
        df_output[column] = df_output[column].map(lambda x: x.replace(',', '.')).astype('float')
        
    df_output['WORK_TIME'].fillna(df_output['WORK_TIME'].mean())
    category_list = 'GEN_INDUSTRY,GEN_TITLE,ORG_TP_STATE,ORG_TP_FCAPITAL,JOB_DIR,TP_PROVINCE,REGION_NM'.split(',')
    for column in category_list:
        df_output[column].fillna('Не определено', inplace=True)
    
    df['PREVIOUS_CARD_NUM_UTILIZED'].fillna(0, inplace=True)
    
    dummy_list = list()
    for c in df_output.columns:
        if c == 'sample':
            continue
        if str(df_output[c].dtype) == 'object':
            dummy_list.append(c)
            
        if str(df_output[c].dtype).startswith('float'):
            df_output[c].fillna(0, inplace=True)
    
    df_output = pd.get_dummies(df_output, columns=dummy_list)
    return df_output
```

```python
df_preproc = df.pipe(preproc_data)

df_train_preproc = df_preproc.query('sample == "train"').drop('sample', axis=1)
df_test_preproc = df_preproc.query('sample == "test"').drop(['sample'], axis=1)
```

#### Задание 5. Обучение и оценка качества разных моделей

```python
from sklearn.cross_validation import train_test_split
# test_size=0.3, random_state=42

## Your Code Here
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
# Постепенно мы узнаем, что они делают а сейчас учимся понимать какой работает качественнее
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

## Your Code Here
#MODELS = (DecisionTreeClassifier, RandomForestClassifier, LogisticRegression)
    
def BIG_BLACK_BOX(MODEL, X_train=X_train, X_test=X_test, y_train=y_train):
    MODEL.fit(X_train, y_train)
    predict = MODEL.predict(X_test)
    predict_proba = MODEL.predict_proba(X_test)
    return MODEL, predict, predict_proba

DTC_MODEL, DTC_predict, DTC_proba = BIG_BLACK_BOX(DecisionTreeClassifier(random_state=42))
RFC_MODEL, RFC_predict, RFC_proba = BIG_BLACK_BOX(RandomForestClassifier(random_state=42))
LR_MODEL, LR_predict, LR_proba = BIG_BLACK_BOX(LogisticRegression(random_state=42))

MODELS = {
    'DTC':DTC_predict,
    'RFC':RFC_predict,
    'LR':LR_predict, 
}

```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall

from sklearn.metrics import accuracy_score, precision_score, recall_score

## Your Code Here
for name, MODEL in MODELS.items():
    print(name)
    print('accurancy_score={:.3f}'.format(accuracy_score(y_test, MODEL)))
    print('precision_score={:.3f} '.format(precision_score(y_test, MODEL)))
    print('recall_score={:.3f}'.format(recall_score(y_test, MODEL)))
    print('')
```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально сравнить)
# Возможно вас удивит качество! Но задача подобрана специально ;) Такое качество тоже бывает

## Your Code Here
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import precision_recall_curve

plt.figure(figsize=(10, 10))
LEGEND = list()
for model_name, MODEL in MODELS.items():
    precision, recall, thresholds = precision_recall_curve(y_test, MODEL)
    plt.plot(precision, recall)
    LEGEND.append(model_name)
    
    
plt.title('precision_recall_curve')
plt.legend(LEGEND)
```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе
# Подпишите оси и линии

## Your Code Here
plt.figure(figsize=(10, 10))
LEGEND = list()
for model_name, MODEL in MODELS.items():
    fpr, tpr, thresholds = roc_curve(y_test, MODEL)
    plt.plot(fpr, tpr)
    LEGEND.append(model_name)
    print('roc_auc_score, model: {}, score: {:.3f}'.format(model_name, roc_auc_score(y_test, MODEL)))
    
plt.title('roc_curve')
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.legend(LEGEND)

```

```python
from sklearn.cross_validation import cross_val_score
# Сделать k-fold (10 фолдов) кросс-валидацию каждой модели
# И посчитать средний roc_auc

## Your Code Here
MODELS = {
    'DTC': DecisionTreeClassifier(random_state=42),
    'RFC': RandomForestClassifier(random_state=42),
    'LR': LogisticRegression(random_state=42)
}
for model_name, MODEL in MODELS.items():
    cvs = cross_val_score(
        MODEL,
        X,
        y,
        scoring='roc_auc',
        cv=10
    )
    print('{}\ncvs_mean={:.3f}\n{}\n'.format(model_name, cvs.mean(), cvs))
```
